```python
from langchain_openai import ChatOpenAI, OpenAIEmbeddings
```

```python
llm_ques_gen_pipeline = ChatOpenAI(
    model='gpt-4o-mini',
    temperature=0.3
)
```

```python
prompt_template = """
You are an expert at creating questions based on the provided material and documentation.
Your goal is to prepare a student or professional for their exam or interview.
You do this by asking questions about the text below:

------------
{text}
------------

Create questions that will prepare the reader for tests or interviews.
Make sure not to lose any important information.

QUESTIONS:
"""
```

```python
from langchain_core.prompts import PromptTemplate, ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnablePassthrough
```

```python
PROMPT_QUESTIONS = PromptTemplate(template=prompt_template, input_variables=['text'])
```

```python
refine_template = ("""
You are an expert at creating practice questions based on the provided material.
Your goal is to help someone prepare for an exam or interview.

Here are the existing questions generated so far:
{existing_answer}

Below is additional context from the document:
------------
{text}
------------

Instructions:
- Add new questions that cover important topics NOT already addressed.
- Remove duplicate or overly similar questions.
- Ensure all questions are clear, specific, and test real understanding.
- Do NOT include questions answerable with a simple yes/no.
- Return the complete final list of questions.

QUESTIONS:
""")
```

```python
REFINE_PROMPT_QUESTIONS = PromptTemplate(
    input_variables=["existing_answer", "text"],
    template=refine_template,
)
```

```python
# Refine chain implemented manually using LCEL (load_summarize_chain removed in LangChain 1.x)
initial_chain = PROMPT_QUESTIONS | llm_ques_gen_pipeline | StrOutputParser()
refine_chain = REFINE_PROMPT_QUESTIONS | llm_ques_gen_pipeline | StrOutputParser()
```

```python
ques = initial_chain.invoke({"text": document_ques_gen[0].page_content})
for doc in document_ques_gen[1:]:
    ques = refine_chain.invoke({"existing_answer": ques, "text": doc.page_content})
```

```python
from langchain_community.vectorstores import FAISS
```

```python
embeddings = OpenAIEmbeddings()
```

```python
# FAISS already imported above from langchain_community.vectorstores
```

```python
vector_store = FAISS.from_documents(document_answer_gen, embeddings)
```

```python
llm_answer_gen = ChatOpenAI(temperature=0.1, model="gpt-4o-mini")
```

```python
ques
```

```python
ques_list = ques.split("\n")
```

```python
ques_list
```

```python
qa_prompt = ChatPromptTemplate.from_template(
    "Answer the question based on the context below.\n\nContext: {context}\n\nQuestion: {question}"
)

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)
```

```python
answer_generation_chain = (
    {"context": vector_store.as_retriever() | format_docs, "question": RunnablePassthrough()}
    | qa_prompt
    | llm_answer_gen
    | StrOutputParser()
)
```

```python
for question in ques_list:
    print("Question: ", question)
    answer = answer_generation_chain.invoke(question)
    print("Answer: ", answer)
    print("--------------------------------------------------\n\n")
    with open("answers.txt", "a") as f:
        f.write("Question: " + question + "\n")
        f.write("Answer: " + answer + "\n")
        f.write("--------------------------------------------------\n\n")
```
