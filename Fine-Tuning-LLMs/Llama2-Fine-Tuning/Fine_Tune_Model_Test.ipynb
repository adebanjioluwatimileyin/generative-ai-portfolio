## Load Fine-Tuned Llama 2 and Run Inference

```python
import torch
from transformers import pipeline

model_id = "entbappy/Llama-2-7b-chat-finetune"

pipe = pipeline(
    "text-generation",
    model=model_id,
    torch_dtype="auto",
    device_map="auto",
)
```

### Example 1 — Basic Question

```python
prompt = "What is a large language model?"

result = pipe(
    f"<s>[INST] {prompt} [/INST]",
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    max_new_tokens=200,
)
print(result[0]["generated_text"])
```

### Example 2 — Multi-Step Explanation

```python
prompt = (
    "Explain the equation for the nth triangle number and prove it "
    "using only high school level math. Give each proof step in LaTeX."
)

result = pipe(
    f"<s>[INST] {prompt} [/INST]",
    do_sample=True,
    top_k=10,
    num_return_sequences=1,
    max_new_tokens=400,
)
print(result[0]["generated_text"])
```
